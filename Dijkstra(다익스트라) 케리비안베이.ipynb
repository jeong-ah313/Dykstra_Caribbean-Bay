```python
select_node=[]
not_going_node=['타워빌리지','타워부메랑고','타워래프트','교차점1','와일드리버풀','교차점2','와일드리버','아쿠아루프','흡연구역1','래프트빌리지','교차점3','키디풀','키디빌리지','스파빌리지','아쿠아틱센터','바하마','라커','산후앙','봅슬레이빌리지','어드벤처빌리지','교차점5','서핑라이드','교차점6','미라클스파','흡연구역2','교차점4','흡연구역3','교차점7','유수풀','어드벤처풀','흡연구역4','캐슬빌리지','메가스톰','교차점8','슬라이드빌리지','카리브스테이지','흡연구역5','웨이브빌리지1','종합대여소','씨웨이브','웨이브빌리지2','교차점9','자켓대여소','라코스타','실외라커동']
start_from_distance=[]
just_before_node=[]
for i in range(45):
    start_from_distance.append(999999)
    just_before_node.append(-1)
```

```python
start=input("출발지를 입력하세요:")
start_index=not_going_node.index(start)  #출발지의 index를 변수에 저장
select_node.append(start)  #출발지를 선택된 노드에 추가
not_going_node.remove(start)  #출발지를 가지 않은 노드에서 제거
start_from_distance[start_index]=0  #출발지부터 출발지의 거리니까 0으로
```

```python
while 1:
    A=0
    B=list(df.columns).index(start)
    for i in list(df.columns):        
        C=list(df.columns).index(i)
        if i in not_going_node:
            if df.loc[start,i]<10000:
                if start_from_distance[B] + df.loc[start,i] < start_from_distance[C]:
                    start_from_distance[C] = start_from_distance[B] + df.loc[start,i]
                    
            else:
                if df.loc[start,i] < start_from_distance[C]:
                    start_from_distance[C] = df.loc[start,i]
    
    MIN1=10000
    for i in list(df.columns):
        C=list(df.columns).index(i)
        if i in not_going_node:
            if start_from_distance[C]<MIN1:
                MIN1 = start_from_distance[C]
                start=i
    
    MIN2=10000
    for i in list(df.columns):
        C=list(df.columns).index(i)
        if start_from_distance[C] + df.loc[i,start] < MIN2:
            MIN2 = start_from_distance[C] + df.loc[i,start]
            A=C
    
    just_before_node[list(df.columns).index(start)]=A
    select_node.append(start)
    not_going_node.remove(start)
    
    if not_going_node==[]:
        break

End = input("도착지를 입력하세요:")
End_index=list(df.columns).index(End)

while 1:
    if just_before_node[End_index]==-1:
        break
    else:
        print("%s  -->  %s" %(df.columns[just_before_node[End_index]],df.columns[End_index]))
        End_index=just_before_node[End_index]
```
